## Example of using specim code to extract a spectrum

```python
""" Import the necessary modules"""
from matplotlib import pyplot as plt
%matplotlib inline

from specim.specfuncs import spec2d
```

```python
"""
Set the input information. This is for one of the exposures on the standard star (G191B2B)
"""
#infile = '../lsr190303_0041r_G191B2B.fits'
infile  = '../0712_G0594_001.fits'
```

```python
""" Load the spectrum into a Spec2d object """
spec = spec2d.Spec2d(infile)
```

```python
""" Make a quick diagnostic plot of the spectrum """
spec.display_spec()
```

```python
"""
Make a 'spatial profile' by collapsing the spectrum along the wavelength axis.
In the plot below the "spatial direction" axis corresponds to the y axis in the
2d spectra above.  You can see in both that the star is about 20% from one
end of the slit.
"""
profile = spec.spatial_profile()
```

```python
#print(spec.mod0)
test = spec.trace_spectrum(spec.mod0, debug=True)
```

```python
"""
This spatial profile just looks like a simple Moffat or Gaussian profile, so just proceed
with defaults
"""
plt.figure(figsize=(9,5))
spec.find_and_trace()
```

```python
spec.extract()
```

```python
spec.extract(method='modelfit')
```
