```python
from ppxf.ppxf import ppxf
import ppxf.ppxf_util as util
from specim.specfuncs import spec1d
import numpy as np
import matplotlib.pyplot as plt
import glob
from random import sample
import pandas as pd
import seaborn as sn
import pm_veldis_util as vd
from collections import Counter
from keckcode.deimos import deimosmask1d
```

```python
coadded_spectra = deimosmask1d.DeimosMask1d('../galaxy_spectra/0712m4-blue-coadd_latest.fits')
```

```python
specid = [] #np.zeros(coadded_spectra.nspec)
info = coadded_spectra.slitinfo
#print(info)
for i in range(coadded_spectra.nspec):
    spec_id = '%d_%d_%d_%d' % (info['det'][i], info['slitid'][i], info['objid'][i], info['spatloc'][i])
    specid.append(spec_id)
    print(spec_id)
```

## The first one is a spectrum of a galaxy at redshift around 0.296 based on Ca K+H lines

```python
%matplotlib notebook

for i in range(len(specid)):
    print('spec_id : %s' %specid[i])
    coadded_spectra[specid[i]].smooth(7)
    coadded_spectra[specid[i]].mark_lines('abs', z=0.405, usesmooth=True)
#plt.figure(figsize=(50,50))
#lines={'em' : True, 'strongem' : True}, index=specid[i]
```

## The second one seems like a stellar spectra

```python
print('spec_id : %s' %specid[1])
coadded_spectra[specid[1]].smooth(7)
coadded_spectra[specid[1]].mark_lines('abs', z=0, usesmooth=True)
```

## This one seems like a stellar spectra but not sure

```python
print('spec_id : %s' %specid[2])
coadded_spectra[specid[2]].smooth(31)
coadded_spectra[specid[2]].mark_lines('abs', z=0, usesmooth=True)
```
